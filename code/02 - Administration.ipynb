```python
import mlflow
import numpy as np
mlflow.set_tracking_uri('http://server:5000')
```

```python
description = 'testing this'
mlflow.create_experiment(name="testing", tags={'user': 'john doe', 'deparment': 'health', 'mlflow.note.content': description})
```

```python
experiments = mlflow.search_experiments(order_by=['last_update_time ASC'])
#experiments = mlflow.search_experiments(filter_string="tags.deparment = 'health'", order_by=['last_update_time DESC'])
print(experiments)
```

# Register best run

```python
experiment_id = experiments[0].experiment_id
runs = mlflow.search_runs(experiment_ids=[experiment_id], order_by=['metrics.mean_absolute_error_X_val'])
runs
```

```python
model_path = f"runs:/{runs.iloc[0].run_id}/model"
model_name = "DiabetesRegression"
model_version = mlflow.register_model(model_uri=model_path, name=model_name)

# Print the model registration details
print(f"Model registered with name: {model_version.name}")
print(f"Model registered with version: {model_version.version}")
```

```python
model_version
```

# Deploy register model

```python
model_uri = f"models:/{model_version.name}/{model_version.version}"
model = mlflow.pyfunc.load_model(model_uri)
```
